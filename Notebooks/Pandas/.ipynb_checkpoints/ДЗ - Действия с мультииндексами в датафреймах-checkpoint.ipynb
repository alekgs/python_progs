## Задача 6
### Загрузите данные из task5.csv. По всей Москве рассчитайте:
### -Общее количество бесплатных парковок,
### -Общее количество платных парковок,
### -Общее количество мест на бесплатных парковках,
### -Общее количество мест на платных парковках

### Введите в поле с ответом получившиеся значения через пробел. В ответе должны быть целые числа
### Пример: 10 12 90 110

```python
df6_task = pd.read_csv('task5.csv', index_col=[0, 1, 2], header=[0, 1]) 
```

```python
df6_task
```

#### Общее количество бесплатных парковок


```python
df6_task['Capacity'].groupby(level=2).sum().loc[['Бесплатная']]['count'][0]
```

#### Общее количество платных парковок

```python
df6_task['Capacity'].groupby(level=2).sum().loc[['Платная']]['count'][0]
```

#### Общее количество мест на бесплатных парковках

```python
df6_task['Capacity'].groupby(level=2).sum().loc[['Бесплатная']]['sum'][0]
```

#### Общее количество мест на платных парковках

```python
df6_task['Capacity'].groupby(level=2).sum().loc[['Платная']]['sum'][0]
```

#### Решение в одну строку (Роман Ковков)

```python
print(*df6_task['Capacity'].groupby(level=2).sum().T.stack().to_list())
```

```python
print(*df6_task['Capacity'].groupby(level=2).sum().unstack().to_list())
```

### Ответ: 518 529 3455 3642
