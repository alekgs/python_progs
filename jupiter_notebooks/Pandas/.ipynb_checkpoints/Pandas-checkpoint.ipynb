### 5.3.6 
На вход функции подается датафрейм с наиболее популярными веб-сайтами за последний месяц в мире. У некоторых веб-сайтов не верно записано доменное имя: вместо .com -> .cm или .om, а вместо .ru -> рф. По всей видимости данные в систему вносил человек, который не очень знаком с интернетом. Замените значения в ячейках на корректные:

```python
df_ = pd.DataFrame({'Веб-сайт': ['google.com', 'youtube.com', 'facebook.cm',
                  'twitter.com', 'instagram.com', 'baidu.com', 'wikipedia.org',
                   'yandex.рф', 'yahoo.cm', 'whatsapp.om']})


def solution(df: pd.DataFrame) -> pd.DataFrame:
    """Принимает dataframe (список веб-сайтов) и заменяет в нем значения .cm или .om на .com, рф на .ru"""
    return df.replace({'\.[co]m': '.com', '\.рф': '.ru'}, regex=True)
    #return df.replace(['\.[co]m$', '\.рф$'], ['.com', '.ru'], regex=True)

    
solution(df_)
```

```python
df = pd.DataFrame([[0, np.nan, np.nan, 3, 4, 5, 6, 7, 8, np.nan],
                    [np.nan, 11, np.nan, 13, 14, 15, 16, 17, 18, np.nan],
                    [np.nan, np.nan, 22, 23, 24, 25, 26, 27, 28, np.nan],
                    [30, 31, 32, 33, 34, np.nan, 36, 37, 38, np.nan],
                    [40, 41, np.nan, 43, 44, 45, 46, 47, 48, np.nan],
                    [50, 51, 52, np.nan, 54, 55, np.nan, 57, 58, np.nan],
                    [60, 61, 62, 63, 64, np.nan, 66, 67, np.nan, np.nan],
                    [np.nan, 71, 72, 73, 74, 75, 76, 77, 78, np.nan],
                    [80, 81, 82, 83, 84, 85, np.nan, 87, 88, np.nan],
                    [90, 91, 92, 93, 94, 95, 96, 97, 98, np.nan]],
                   columns=["A", "B", "C", "D", "E", "F", "G", "H", "J", "K"])
df
```

```python
# Столбец А: 0
df['A'].fillna(0,inplace=True)

# Столбец B: среднее значение из столбца Е
df['B'].fillna(df['E'].mean(),inplace=True)

# Столбец С: максимальное значение из столбца H
df['C'].fillna(df['H'].max(),inplace=True)

# Столбец F: по методу ffill
df['F'].ffill(inplace=True)

# Столбец G: по методу bfill
df['G'].bfill(inplace=True)

# Удалите столбец в котором все значения пропущены. Функцию drop не использовать!
df.dropna(axis=1, how='all', inplace=True)

# Удалите строки, в которых хотя бы одно значение пропущено
df.dropna(inplace=True)
```

### Создайте новый столбец таким образом, чтобы каждая строка содержала номер наиболее удалённой строки-записи по евклидову расстоянию


Переводите данные в np.array;
Находите векторную разницу в квадрате между np.array и np.array с новой размерностью (np.newaxis в помощь). 
На выходе массив размерностью (10, 10, 4).
Суммируете по последней размерности и берете корень. 
Получаете массив 10, 10 c дистанцией между всеми строками. 
Если все сделали до этого правильно, то на главной диагонали будут нули, что соответствует расстоянию строки до себя самой же.
Дальше применяя сортировку и метод np.argsort можете вытащить значения дистанций и номера! индексов строк;
Останется поиграть с типом данных для дистанции и округлением, добавляем данные в df


```python
df = pd.DataFrame(np.random.RandomState(30).randint(1,100, 40).reshape(10, -1), columns=list('pqrs'), index=list('abcdefghij'))
print(df)

def solution(df):
    temp = df.apply(lambda x: (df-x).pow(2).sum(axis=1).pow(0.5), axis=1)
    df['nearest_row'] = temp.idxmax()
    df['dist'] =  temp.max().round().astype('int')
    return df
```

```python
solution(df)
```
